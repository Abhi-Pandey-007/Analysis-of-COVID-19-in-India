## Data analysis and visualization

### Relating the variables with scatterplots

```python
sns.relplot(x="Confirmed",y="Cured",data=data)
plt.ticklabel_format(style='plain')
```

```python
sns.relplot(x="Confirmed",y="Deaths",data=data)
#plt.ticklabel_format(style='plain')
```

From the aforementioned two plots, it is evident that as the number of confirmed cases rises in the country then along with it the number of cured people as well as deaths due to COVID also increases. Thus, an approximate directly proportional realationship between confirmed cases and deaths and confirmed cases and cured cases could be seen and thus envisaged too.

```python
sns.relplot(x="ConfirmedForeignNational",y="ConfirmedIndianNational",data=data)
```

```python
sns.relplot(x="Confirmed",y="State/UnionTerritory",data=data)
```

```python
sns.relplot(x="Deaths",y="State/UnionTerritory",data=data)
```

```python
sns.relplot(x="Cured",y="State/UnionTerritory",data=data)
```

From the aforementioned plots, it is lucid that Maharashtra is the most severely affected state due to COVID, with Kerala occupying the second position.

```python
sns.relplot(x="Confirmed",y="Cured",data=data, hue="Deaths")
plt.ticklabel_format(style='plain')
```

```python
sns.pairplot(data)
```

```python
data.columns
```

```python
sns.relplot(x='Confirmed',y='Deaths',kind='line',data=data)
```

```python
sns.relplot(x='Confirmed',y='Cured',kind='line',data=data)
```

```python
sns.relplot(x="Date",y="Confirmed",data=data, hue="Deaths")
```
